# Evaluaci√≥n de clics y tiempo de partida registrados seg√∫n el usuario

* Usuario Rodrigo

```python
# Cargamos los m√≥dulos
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Leer el archivo CSV sin nombres de columnas
data = pd.read_csv("Rodrigo_11_matches.csv", sep=";", header=None)

# Renombrar columnas
data.columns = ["Nivel", "Usuario", "Clics", "Tiempo"]
```

```python
# Convertir Nivel a num√©rico
data["Nivel"] = data["Nivel"].str.extract(r"(\d+)").astype(int)

# Convertir tiempo a segundos
data["Tiempo"] = pd.to_timedelta(data["Tiempo"]).dt.total_seconds()
```

```python
data_long = pd.melt(data, id_vars=["Nivel"], value_vars=["Clics", "Tiempo"], var_name="M√©trica", value_name="Valor")
```

```python
# Creamos los gr√°ficos
sns.set(style="whitegrid")
g = sns.catplot(
    data=data_long, x="Nivel", y="Valor", hue="M√©trica", col="M√©trica",
    kind="bar", height=4, aspect=1.2, palette="muted", sharey=False
)

# Ajustar el gr√°fico
g.set_titles("{col_name}")
g.set_axis_labels("Niveles", "Valores")
g.set_xticklabels(rotation=45)
g.fig.suptitle("Clics y Tiempo por Nivel del usuario Rodrigo", y=1.02)
plt.show()
```

```python
# Guardar el gr√°fico en formato .PNG
g.savefig("rodrigo_clicks_tiempo.png", dpi=300)
```

```python
# Crear el gr√°fico combinado
fig, ax1 = plt.subplots(figsize=(10, 6))

# Graficar las barras (Clics)
ax1.bar(data["Nivel"], data["Clics"], color="skyblue", label="Clics", alpha=0.7)
ax1.set_xlabel("Niveles")
ax1.set_ylabel("N√∫mero de Clics", color="blue")
ax1.tick_params(axis="y", labelcolor="blue")

# Crear un segundo eje para el Tiempo
ax2 = ax1.twinx()
max_clics = data["Clics"].max()
max_tiempo = data["Tiempo"].max()

# Normalizar el tiempo para ajustarlo al segundo eje
ax2.plot(data["Nivel"], data["Tiempo"] * max_clics / max_tiempo, color="red", label="Tiempo", marker="o")
ax2.set_ylabel("Tiempo (segundos)", color="red")
ax2.tick_params(axis="y", labelcolor="red")

# T√≠tulos y leyenda
fig.suptitle("Clicks y Tiempo por Nivel del usuario Rodrigo")
fig.tight_layout()
ax1.legend(loc="upper left")
ax2.legend(loc="upper right")

# Mostrar el gr√°fico
plt.show()
```
